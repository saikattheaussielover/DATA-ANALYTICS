```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Load job market
df = pd.read_csv("job-market(3).csv")

df.head()
```

```python
%matplotlib inline

job_counts = df['Location'].value_counts()
job_counts.plot.barh(figsize=(12,12))
```

```python
import matplotlib
#create dataframe for salary ranges

job_salary = df['LowestSalary'].value_counts()
job_salary.head()
```

```python
#salary ranges plot
job_salary.plot.pie(figsize=(5,5), autopct='%1.2f%%')
```

```python
#OPTIONAL

locationDf = df[df['Location'].isna()==False]
locationDf = locationDf.groupby('Location').count()
locationDf['Location'] = locationDf.index
locationDf['Count'] = locationDf["Id"]
locationDf.head()
```
